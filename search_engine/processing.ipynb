```python
import chromadb 

client = chromadb.PersistentClient(path="final_project/")
```

```python
original_data = pd.read_csv("/kaggle/input/subtitles-2/subtitles")

chunked_embeddings_batches = []
for i in range(138):  
    filename = f"/kaggle/input/subtitles-2/Chunked_data/chunked_data_batch_{i}.csv"
    try:
        chunked_embeddings_batch = pd.read_csv(filename)
        chunked_embeddings_batches.append(chunked_embeddings_batch)
    except FileNotFoundError:
        print(f"File {filename} not found. Skipping...")

data = []
for idx, batch in enumerate(chunked_embeddings_batches):
    try:
        num = original_data.at[idx, 'num']
        name = original_data.at[idx, 'name']
        for _, row in batch.iterrows():
            data.append([num, name, row['Text'], row['Embedding']])
    except IndexError:
        print(f"No batch found for row {idx} in original_data. Skipping...")

chunked_embeddings_df = pd.DataFrame(data, columns=['num', 'name', 'Text', 'Embedding'])

merged_data = pd.merge(original_data, chunked_embeddings_df, on=['num', 'name'])

```

```python
def add_data_in_batches(collection, documents, embeddings, metadatas, ids, batch_size=1000):
    num_batches = len(documents) // batch_size + 1
    for i in range(num_batches):
        start_idx = i * batch_size
        end_idx = min((i + 1) * batch_size, len(documents))
        batch_documents = documents[start_idx:end_idx]
        batch_embeddings = embeddings[start_idx:end_idx]
        batch_metadatas = metadatas[start_idx:end_idx]
        batch_ids = ids[start_idx:end_idx]
        collection.add(
            documents=batch_documents.tolist(),
            embeddings=batch_embeddings,
            metadatas=batch_metadatas,
            ids=batch_ids
        )

collection = client.get_or_create_collection(name="subtitles")

unique_ids = np.arange(1, len(merged_data) + 1)
merged_data['new_id'] = unique_ids

documents = merged_data['Text']
embeddings = merged_data['Embedding'].tolist()
names = merged_data['name']
ids = merged_data['num'].tolist()

names = names.astype(str)
ids = merged_data['new_id'].astype(str).tolist()

metadatas = [{"num": num, "name": name, "text": text, "embedding": emb} 
             for num, name, text, emb in zip(merged_data['new_id'], names, documents, embeddings)]

batch_size = 1000  

add_data_in_batches(collection, documents, embeddings, metadatas, ids, batch_size=batch_size)
```
